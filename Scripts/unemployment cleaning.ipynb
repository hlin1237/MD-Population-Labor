```python
import pandas as pd

file_path = r"/Users/huiminlin/Downloads/Unemployment.xlsx"

df = pd.read_excel(file_path, skiprows=4)

df
```

```python
df = df[df["State"] == "MD"]

df
```

```python
columns_to_keep = ['FIPS_Code', 'State', 'Area_Name'] + [col for col in df.columns if col.startswith('Unemployed_')]

# Filter the DataFrame
df_filtered = df[columns_to_keep]
df_filtered
```

```python
# Check if there are any NA values in the DataFrame
na_values = df_filtered.isna().sum()

# Display the columns with NA values and their count
print(na_values)
```

```python
new_file_path = r"/Users/huiminlin/Downloads/Cleaned_Unemployment.xlsx"
df_filtered.to_excel(new_file_path)
```
